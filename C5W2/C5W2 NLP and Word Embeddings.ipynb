<br><br>

<br><br>

Man:Computer_Programmer as Woman: Homemaker
<br><br>

Father:Doctor as Mother:Nurse
<br><br>

Word embeddings can reflect gender, ethnicity, age, sexual orientation, and other biases of the text used to train the model.
